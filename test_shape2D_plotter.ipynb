```python
"""testing multiple 2D shapes in one plot"""
from circle import Circle
from rectangle import Rectangle
from shape2D_plotter import Shape2DPlotter

# Create shapes
c1 = Circle(x=0, y=0, radius=2)
c2 = Circle(x=-4, y=-2, radius=1.5)
r1 = Rectangle(x=4, y=2, width=3, height=2)
r2 = Rectangle(x=-3, y=3, width=2, height=1)

# Create plotter
plotter = Shape2DPlotter()

# Add shapes
plotter.add_circle(c1, color='blue')
plotter.add_circle(c2, color='orange')
plotter.add_rectangle(r1, color='green')
plotter.add_rectangle(r2, color='red')

# Show all shapes together
plotter.show()
```
